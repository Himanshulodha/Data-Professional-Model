# Model Evolution 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
print(df_modified_with_salary.columns)
```

```python
# Initialize an empty dictionary to store LabelEncoder objects
label_encoders = {}

# Iterate through the columns you want to label encode
for column in ['Data Roles', 'Industry', 'Programming Language', 'Difficult for breaking Data', 'Things to look new job', 'Sex', 'Country', 'Ethnicity']:
    # Initialize a LabelEncoder object for each column and fit-transform the column data
    label_encoders[column] = LabelEncoder()
    df_modified_with_salary[column] = label_encoders[column].fit_transform(df_modified_with_salary[column])
```

```python
# Splitting the data into features and target variable
X = df_modified_with_salary.drop(columns=['Average Salary', 'Salary Range'])
y = df_modified_with_salary['Average Salary']
```

```python
# Exclude non-numeric columns 
numeric_columns = df_modified_with_salary.select_dtypes(include=['float64', 'int64']).columns
X = df_modified_with_salary[numeric_columns]

#Convert non-numeric value to numeric
#This code selects only the columns with numeric data types (float64 and int64) and applies feature scaling to them. This should resolve the ValueError caused by non-numeric values in the dataset.
```

```python
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Choose and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Fill missing values with mean
df_modified_with_salary.fillna(df_modified_with_salary.mean(), inplace=True)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_test_imputed = imputer.fit_transform(X_test)

# Feature scaling
scaler = StandardScaler()
X_test_scaled = scaler.fit_transform(X_test_imputed)
```

```python
# Making predictions
y_pred_test = model.predict(X_test_scaled)
```

```python
y_pred_train = model.predict(X_train)
```

```python
# Evaluating the model
train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
train_r2 = r2_score(y_train, y_pred_train)
```

```python
print("Training RMSE:", train_rmse)
# print("Testing RMSE:", test_rmse)
print("Training R^2 Score:", train_r2)
# print("Testing R^2 Score:", test_r2)
```
